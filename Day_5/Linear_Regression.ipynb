```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('salary_dataset.csv', index_col= 0)
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(x=data['YearsExperience'], y=data['Salary'])
```

```python
x= data['YearsExperience']
x.head()
```

```python
y= data["Salary"]
y.head()
```

```python
x = x.values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8 , random_state=2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)     
```

$Y= \theta_0 + \theta_1 \times Y.E$

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```

```python
y_test.shape
```

```python
c = [i for i in range(1,7,1)]

plt.figure()
plt.plot(c,y_test, color="black", label = 'actual')
plt.plot(c,y_pred, color="red", label = 'predicted')
plt.title('Actual and Predicted', fontsize=15)              
plt.xlabel('Index')                               
plt.ylabel('Sales')           
plt.legend(); 
```

```python
c = [i for i in range(1,7,1)]

plt.figure()
plt.plot(c,y_test-y_pred, color="blue",)
plt.title('Error Terms', fontsize=15)              
plt.xlabel('Index')                     
plt.ylabel('ytest - ypred');             
```
